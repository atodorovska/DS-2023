#### <font color='white'>Handling of missing values - KNN Based Imputation<font>


**KNN Based Imputation** follows the following steps: (uses Classification)
1. Setting the value of k
2. Normalization
3. Calculating the Euclidean distance
4. Prediction

```python
#Note: Step 1 - Setting the value of k

k = 2
```

```python
#Note: Step 2 - Normalization

from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
df_copy = df.copy()
```

```python
df_copy['LotFrontage']
```

```python
type(df_copy['LotFrontage'])
```

```python
#Note: In NumPy, -1 in reshape(-1) refers to an unknown dimension that the reshape() function calculates for you.
#Resource: https://www.codingem.com/numpy-reshape-minus-one/#:~:text=In%20NumPy%2C%20%2D1%20in%20reshape,elements%20to%20a%201D%20array.

df_copy['LotFrontage'].to_numpy().reshape(-1, 1)
```

```python
df_copy['LotFrontage'] = scaler.fit_transform(df_copy['LotFrontage'].to_numpy().reshape(-1, 1))
```

```python
df_copy['LotFrontage'].head(20)
```

```python
#Note: Step 3 - Calculating the Euclidean distance & Step 4 - Prediction

from sklearn.impute import KNNImputer
```

```python
#Note: With the n_neighbors we can actually set the k parameter i.e. the number of closest neighbours we look at. The default value is 5.

imputer = KNNImputer(n_neighbors=k)
```

```python
df_copy['LotFrontage'] = imputer.fit_transform(df_copy['LotFrontage'].to_numpy().reshape(-1, 1))
```

```python
df_copy['LotFrontage'].head(20)
```

```python
df_copy['LotFrontage'].isnull().sum()
```

#### <font color='white'>Handling of missing values - MICE Imputation<font>



**MICE** uses Linear regression and the Round Robin algorithm. It divides the data into x chunks and uses Linear regression on each chunk. With the results that are obtained, an average is calculated and put in all the places where a NaN value was present.

```python
#Note: Use the following documentation to see features and example usages.
#Resource: https://scikit-learn.org/stable/

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
#Resource: https://scikit-learn.org/stable/modules/generated/sklearn.impute.IterativeImputer.html

iterative_imputer = IterativeImputer(random_state=0)
```

```python
df_copy = df.copy()
```

```python
df_copy['LotFrontage'] = iterative_imputer.fit_transform(df_copy['LotFrontage'].to_numpy().reshape(-1, 1))
```

**Forward Fill**

```python
#Note: Forward Fill means filling in with the last valid data, which is propagated forward.
#Note: You can also use the ffill() method.

df_copy = df.copy()
df_copy.fillna(method='ffill', inplace=True)
```

```python
df_copy['Xylene'][50:70]
```

**Backward Fill**

```python
#Note: Backward Fill means filling in with the first next valid data, which is propagated backward.
#Note: You can also use the bfill() method.

df_copy = df.copy()
df_copy.fillna(method='bfill', inplace=True)
```

```python
df_copy['Xylene'][50:70]
```

**Linear Interpolation** takes two points of data and performs the formula. It uses all the features dimensions for these points.

image.png

```python
#Note: The 'both' attribute in the interpolation means that it will look at the one point before and one point after the missing data point.

df_copy = df.copy()
df_copy.interpolate(limit_direction='both', inplace=True)
```

```python
df_copy['Xylene'][50:70]
```

```python
#Note: This is the fitting part.

dict = {}
classes = list(set(df_copy['FireplaceQu']))

for i in range(0, len(classes)):
    dict[classes[i]] = i

dict
```
